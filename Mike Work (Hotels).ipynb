```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from pprint import pprint
import datetime
import os
import gmaps
from config import gkey
gmaps.configure(api_key=gkey)
```

```python
hotel = pd.read_csv("Resources/Datafiniti_Hotel_Reviews.csv")
hotel.head()
```

```python
hotel.province.value_counts()
```

```python
locations = hotel[["latitude", "longitude"]].astype(float)
ratings = hotel["reviews.rating"]
```

```python
figure_layout = {
    'width': '640px',
    'height': '640px',
    'border': '1px solid black',
    'padding': '1px',
    'maptype': 'satellite'
   
}

gmaps.figure(layout=figure_layout, display_toolbar=False, tilt=45)

fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=ratings, 
                                 dissipating=False, max_intensity=50,
                                 point_radius = 1)
fig.add_layer(heat_layer)

fig
```
